```python
input_file = 'C:/Users/nicol/OneDrive/Desktop/retail_data/store_branches.csv'
output_file = 'C:/Users/nicol/OneDrive/Desktop/retail_data/store_branches_utf8_clean.csv'

with open(input_file, 'r', encoding='windows-1252', errors='ignore') as src, \
     open(output_file, 'w', encoding='utf-8') as dst:
    for line in src:
        clean_line = ''.join(char if ord(char) < 128 or char.isprintable() else '?' for char in line)
        dst.write(clean_line)
```
